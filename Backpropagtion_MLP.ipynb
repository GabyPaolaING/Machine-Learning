# Backpropagtion MLP

#1.Fundamentos y matemáticos de una Red Neuronal Backpropagation:


Una Red Neuronal Artificial es un modelo inspirado en el funcionamiento del cerebro humano.

Está formada por neuronas artificiales organizadas en capas:


*   Capa de entrada (Input Layer)
*   Capas ocultas (Hidden Layer)
*   Capa de salida (Output Layer)

Cada neurona realiza una operacion matemática:

z = w*x + b

donde:

w = pesos
x = entradas
b = bias

Después se aplica una función de activación:

a = f(z)

Esto permite que la red neuronal pueda aprender relaciones complejas entre los datos.

## 1.1 Forward Pass

El Forward Pass es el proceso donde los datos viajan desde la entrada hasta la salida de la red neuronal.

En cada capa se realizan tres operaciones:


1.   Multiplicación de entradas por pesos
2.   Suma del bias
3.   Aplicación de función de activación

Matemáticamente:

Z = XW + b

A = f(Z)

Donde:

X = datos de entrada  
W = pesos  
b = bias  
A = activación

ya que este proceso continúa hasta generar la predicción final de la red neuronal.




## 1.2 Backward Pass (Backpropagation)

El Backpropagation es el algoritmo que permite que la red neuronal aprenda de sus errores.

Primero se calcula el error entre el valor real y la predicción:

Error = Valor real - Predicción

Después se calcula el gradiente del error respecto a los pesos utilizando la regla de la cadena.

Los pesos se actualizan con la siguiente fórmula:

W = W - learning_rate * gradiente

Este proceso se repite muchas veces hasta que el error se reduce y la red aprende el patrón de los datos.

## 5.1 Model Selection


La razón es que los datos no pueden separarse con una sola línea en el espacio de características.

Por esta razón se utiliza una **Red Neuronal con al menos una capa oculta**, ya que este tipo de modelo puede aprender **relaciones no lineales** entre los datos.

Las redes neuronales utilizan funciones de activación no lineales y múltiples capas para transformar los datos de entrada y encontrar patrones complejos.

##5.2 Model Training

```python
import numpy as np  #importamos la libreria NumPy
```

##Dataset


```python
#Cada fila es un entrenamiento
# Datos de entrada XOR
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Salidas esperadas
y = np.array([
    [0],
    [1],
    [1],
    [0]
])
```

##Función de activación

```python
#Definimos la función sigmoid
#Esta función transforma cualquier valor en un rango entre 0 y 1
def sigmoid(x):
    return 1/(1 + np.exp(-x))

#Calcula como deben ajustarse los pesos

def sigmoid_derivative(x):
    return x * (1 - x)
```

##Inicializamos los pesos

```python
#la semilla para los numeros aleatorios
np.random.seed(1)

#Pesos entre la capa de entrada y la capa oculta
weights_input_hidden = np.random.rand(2,2)

#Pesos entre la capa oculta y la capa de salida
weights_hidden_output = np.random.rand(2,1)

#Bias de la capa oculta
bias_hidden = np.random.rand(1,2)

#Bias de la capa de salida
bias_output = np.random.rand(1,1)
```

##Entrenamiento

```python
#Parametros de entrenamiento
learning_rate = 0.1 # Este controla que tan rapido aprende la neurona
epochs = 10000  #Numero de interaciones de entrenamiento

#Proceso de entrenamiento
for epoch in range(epochs):

    # Forward pass

    #Calculamos la entreda de la capa oculta
    hidden_input = np.dot(X, weights_input_hidden) + bias_hidden

    #Aplicamos la función de activación
    hidden_output = sigmoid(hidden_input)

    # Calculamos la entrada de la capa de salida
    final_input = np.dot(hidden_output, weights_hidden_output) + bias_output


    # Aplicamos nuevamente la función de activación
    predicted_output = sigmoid(final_input)

    # Error
    error = y - predicted_output

    # Backpropagation
    # Calculamos el gradiente del error en la salida
    d_predicted_output = error * sigmoid_derivative(predicted_output)

     # Propagamos el error hacia la capa oculta
    error_hidden = d_predicted_output.dot(weights_hidden_output.T)

    # Calculamos el gradiente para la capa oculta
    d_hidden_layer = error_hidden * sigmoid_derivative(hidden_output)

    # Actualización de pesos
     # Ajustamos los pesos entre la capa oculta y la salida
    weights_hidden_output += hidden_output.T.dot(d_predicted_output) * learning_rate

    # Ajustamos los pesos entre la entrada y la capa oculta
    weights_input_hidden += X.T.dot(d_hidden_layer) * learning_rate

    # Ajustamos los bias de la capa de salida
    bias_output += np.sum(d_predicted_output, axis=0) * learning_rate

     # Ajustamos los bias de la capa oculta
    bias_hidden += np.sum(d_hidden_layer, axis=0) * learning_rate
```

##Prediction

```python
def predict(input_data):

    # Forward pass hacia la capa oculta
    hidden = sigmoid(np.dot(input_data, weights_input_hidden) + bias_hidden)

    # Forward pass hacia la capa de salida
    output = sigmoid(np.dot(hidden, weights_hidden_output) + bias_output)

    return output
```

```python
for x in X:
    print(x, "->", predict(x))
```
