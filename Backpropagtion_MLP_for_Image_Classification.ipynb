#Multi-layer Perceptron for Image Classificaction


##1.Descripción del problema de clasificación por enfrentar.


En este tutorial se aborda un problema de clasificación supervisada utilizando una Red Neuronal Multicapa (MLP).

Se utilizará el dataset Iris, el cual contiene información de flores y busca clasificarlas en tres tipos:

*  Setosa

*  Versicolor

*  Virginica

Cada registro contiene 4 características:

*  Largo del sépalo

*  Ancho del sépalo

*  Largo del pétalo

*  Ancho del pétalo

Su objetivo es que apartir de estas caracteristicas, prediga correctamente la especie de la flor.


##2. Pipeline


2.1.1 Describir las variables

```python
import pandas as pd
from sklearn.datasets import load_iris

# Cargar el dataset Iris directamente desde sklearn
iris = load_iris()

# Crear un DataFrame con los datos de las características y los nombres de las columnas
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Añadir la columna de la clase (variable dependiente)
df['class'] = iris.target_names[iris.target]

# Mostrar primeras filas
df.head()
```

Interpretacion de variables:

*  sepal_length: Largo del sépalo (numérico)

*  sepal_width: Ancho del sépalo (numérico)

*  petal_length: Largo del pétalo (numérico)

*  petal_width: Ancho del pétalo (numérico)

*  class: Tipo de flor (categórico)

2.1.2 Variables independientes y dependiente

```python
import pandas as pd

# Variables independientes (entrada)
X = df.iloc[:, 0:4]

# Variable dependiente (salida)
y = df.iloc[:, 4]

print("Variables de entrada (X):")
print(X.head())

print("\nVariable de salida (y):")
print(y.head())
```

Explicación:

*  X = características (datos de entrada)

*  y = etiqueta (lo que queremos predecir)

2.1.3 Estandarización del dataset

¿Por qué estandarizar?

*  Las redes neuronales trabajan mejor con datos en la misma escala

*  Mejora la velocidad de entrenamiento

*  Evita que una variable tenga más peso que otra

```python
#ejemplo:
from sklearn.preprocessing import StandardScaler

# Crear objeto escalador
scaler = StandardScaler()

# Ajustar y transformar los datos
X_scaled = scaler.fit_transform(X)

print("Datos originales:")
print(X.head())

print("\nDatos estandarizados:")
print(X_scaled[:5])
```

Después de esto:

*  Media ≈ 0

*  Desviación estándar ≈ 1

##2.2 Model Selection

¿Qué es una Red Neuronal MLP?

Una Red Neuronal Multicapa (MLP) es un modelo inspirado en el cerebro humano.

Está formada por:

*  Capa de entrada (datos)

*  Capas ocultas (procesamiento)

*  Capa de salida (resultado)

Cada neurona recibe información, la procesa y la pasa a la siguiente capa.

El aprendizaje ocurre mediante un algoritmo llamado Backpropagation, que ajusta los pesos para reducir el error.

Division de datos:


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

Definicion del modelo MLP

```python
from sklearn.neural_network import MLPClassifier

# Crear modelo
mlp = MLPClassifier(
    hidden_layer_sizes=(10, 10),  # dos capas ocultas
    activation='relu',            # función de activación
    solver='adam',                # optimizador
    max_iter=1000,                # iteraciones
    random_state=42
)
```

*  hidden_layer_sizes: número de capas y neuronas

*  activation: función de activación

*  solver: algoritmo de optimización

*  max_iter: número de épocas

##2.3 Model Training

```python
# Entrenar modelo
mlp.fit(X_train, y_train)

print("Modelo entrenado correctamente")
```

##2.4 Model Prediction

```python
#Funcion de prediccion:
def predecir_flor(modelo, scaler, datos):
    # Escalar datos nuevos
    datos_escalados = scaler.transform([datos])

    # Realizar predicción
    prediccion = modelo.predict(datos_escalados)

    return prediccion[0]
```

```python
#Ejemplo de prediccion

ejemplo = [5.1, 3.5, 1.4, 0.2]

resultado = predecir_flor(mlp, scaler, ejemplo)

print("Predicción:", resultado)
```

##2.5 Model Evaluation

```python
#Matriz de confusion
from sklearn.metrics import confusion_matrix

# Predicciones
y_pred = mlp.predict(X_test)

# Matriz
matriz = confusion_matrix(y_test, y_pred)

print("Matriz de confusión:")
print(matriz)
```

Filas → valores reales

Columnas → valores predichos

*  Los valores en la diagonal = aciertos
*  Los demás valores = errores

```python
#Reporte de clasificacion
from sklearn.metrics import classification_report

reporte = classification_report(y_test, y_pred)

print("Reporte de clasificación:")
print(reporte)
```

*  Precision → qué tan exacto es el modelo

*  Recall → qué tanto detecta correctamente

*  F1-score → balance entre ambos

```python
#este indica el porcentaje de aciertos del modelo, predicciones correctas
accuracy = mlp.score(X_test, y_test)

print("Accuracy del modelo:", accuracy)
```
